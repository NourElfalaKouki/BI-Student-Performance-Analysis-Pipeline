**Bronze layer**

This notebook documents the ingestion of the 9 CSV source files using the Databricks Data UI.
Each source file is now a Delta table in the **bronze layer**, ready for cleaning and transformation.


```python
# Lister les tables dans le schéma par défaut
tables = spark.sql("SHOW TABLES IN default")
display(tables)
```

```python
expected_tables = [
    "student_info", "student_vle", "assessments", "student_assessment",
    "student_registration", "vle", "courses", "online_classroom_data"
]

for table in expected_tables:
    exists = spark.sql(f"SHOW TABLES IN default LIKE '{table}'").count() > 0
    status = "✅ FOUND" if exists else " MISSING"
    print(f"{status} - {table}")

```

```python
for table in expected_tables:
    if spark.sql(f"SHOW TABLES IN default LIKE '{table}'").count() > 0:
        print(f"\n {table.upper()} - SAMPLE:")
        display(spark.table(f"default.{table}").limit(3))
```
